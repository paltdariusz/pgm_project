```python
# Imports
from ydata_profiling import ProfileReport
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
raw_df = pd.read_csv("../data/weatherHistory.csv")

raw_df['Formatted Date'] = pd.to_datetime(raw_df['Formatted Date'], utc=True, errors='coerce')
raw_df.set_index('Formatted Date', inplace=True)

raw_df = raw_df.sort_values(by='Formatted Date')
raw_df.head()
```

```python
raw_df.info()
```

```python
raw_df.describe()
```

```python
profile = ProfileReport(raw_df, title="WeatherData", explorative=True)
profile.to_file("weather_data_report.html")
```

```python
sns.pairplot(raw_df)
plt.show()
```

# Czyszczenie danych

```python
# drop Loud Cover column bo ma stałą wartość = 0
raw_df.drop(columns=['Loud Cover'], inplace=True)

# dodanie kolumny z temperaturą w Kelvinach
raw_df['Temperature (K)'] = raw_df['Temperature (C)'] + 273.15
```

```python
raw_df.head()
```

## Pomiary temperatury

```python
plt.figure(figsize=(20, 5))
plt.plot(raw_df.index.values, raw_df['Temperature (C)'].values)
```

```python
raw_df.shape
```

## Bierzemy pomiar co 6 godzin

```python
sampled_df = raw_df.resample('6H').first()
sampled_df.head()
```

```python
sampled_df.shape
```

```python
plt.figure(figsize=(20, 5))
plt.plot(sampled_df.index.values, sampled_df['Temperature (C)'].values)
```

## Filtrowanie danych po czasie: 2006-2008

```python
filtered_df = sampled_df['2006-01-01':'2008-12-31']
filtered_df.head()
```

```python
filtered_df.shape
```

```python
plt.figure(figsize=(20, 5))
plt.plot(filtered_df.index.values, filtered_df['Temperature (C)'].values)
plt.title('Temperature in Celsius from 2006 to 2008')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
plt.plot(filtered_df.index.values, filtered_df['Temperature (K)'].values)
plt.title('Temperature in Kelvin from 2006 to 2008')
plt.show()
```

```python
# save filtered data to csv
filtered_df.to_csv('../data/filtered_weather_data.csv')
```
